```python
import requests
```

```python
from bs4 import BeautifulSoup
import json
import re
headers = {'User-Agent': 'Mozilla/5.0 (X11; CrOS i686 2268.111.0) AppleWebKit/536.11 (KHTML, like Gecko) Chrome/20.0.1132.57'}

artist_id = []
artist_name = []
type = []
debue = []
artist_img = []
for Id in range(839533, 839733) :
    url = "https://www.melon.com/artist/timeline.htm?artistId=" + str(Id)
    res = requests.get(url, headers=headers).text
    soup = BeautifulSoup(res, "html.parser")
    if not soup.select('div.wrap_atist_info') :
        continue;
    
    artist_id.append(Id)
    
    name = soup.select('p.title_atist')
        
    name = name[0].text.replace("아티스트명", "").strip()
    artist_name.append(name)
    print(Id)

    type_ = soup.select('div.wrap_atist_info > dl > dd')
    type_name = soup.select('div.wrap_atist_info > dl > dt')

    position = 0
    for i in type_name :
        if i.text == "활동유형" :
            break
        position+=1
    if type_[position] :
        type_ = type_[position].text
    else :
        type_ = []
    type.append(type_)
    
    img = soup.select('#artistImgArea')
    img = img[0].find("img")["src"]
    artist_img.append(img)
```

```python
import pandas as pd
from IPython.display import display

df = {"artist_id": artist_id, "artist_name": artist_name, "type": type, "artist_img": artist_img}
display(pd.DataFrame(df))
```
