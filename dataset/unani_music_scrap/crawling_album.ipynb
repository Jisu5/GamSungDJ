```python
import requests
```

```python
from bs4 import BeautifulSoup
import json
import re
headers = {'User-Agent': 'Mozilla/5.0 (X11; CrOS i686 2268.111.0) AppleWebKit/536.11 (KHTML, like Gecko) Chrome/20.0.1132.57'}

album_id = []
album_name = []
album_rate = []
album_genre = []
singer_id = []
album_intro = []
album_img = []
for Id in range(10077877, 10078177) :

    url = "https://www.melon.com/album/detail.htm?albumId=" + str(Id)
    res = requests.get(url, headers=headers).text
    soup = BeautifulSoup(res, "html.parser")
    if not soup.select('div.wrap_info') :
        continue;

    album_id.append(Id)

    name = soup.select('div.song_name')
    name = name[0].text.replace("앨범명", "").strip()
    album_name.append(name)
    print(Id)
    #가수가 1개가 아닐수도있음 수정필요
    singerId = soup.select('div.artist')
    if singerId[0].find("a") :
        singerId = singerId[0].find("a")["href"]
        singerId = re.findall("\d+",singerId)[0]
    else :
        singerId = singerId[0].text.strip()
    singer_id.append(singerId)

    rate = soup.select('div.meta > dl > dd')
    rate = rate[0].text
    album_rate.append(rate)

    #장르 배열로 바꾸기
    genre = soup.select('div.meta > dl > dd')
    genre_name = soup.select('div.meta > dl > dt')
    position = 0
    for i in genre_name :
        if i.text == "장르" :
            break
        position+=1
    if genre[position] :
        genre = genre[position].text
    else :
        genre = []
    album_genre.append(genre)

    intro = soup.select('div.dtl_albuminfo > div')
    if intro :
        intro = intro[0].text.strip()
    album_intro.append(intro)

    img = soup.select('div.thumb > a')
    img = img[0].find("img")["src"]
    album_img.append(img)

```

```python
import pandas as pd
from IPython.display import display

df = {"album_id": album_id, "album_name": album_name, "album_rate": album_rate, "album_genre": album_genre, "singer_id": singer_id, "album_intro": album_intro, "album_img": album_img}
display(pd.DataFrame(df))
```
