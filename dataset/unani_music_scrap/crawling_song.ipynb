```python
import requests
```

```python
from bs4 import BeautifulSoup
import json
import re
headers = {'User-Agent': 'Mozilla/5.0 (X11; CrOS i686 2268.111.0) AppleWebKit/536.11 (KHTML, like Gecko) Chrome/20.0.1132.57'}

songID = []
song_name = []
singerID = []
albumID = []
song_genre = []
lyric = []
for Id in range(30514366, 30514666) :
    url = "https://www.melon.com/song/detail.htm?songId=" + str(Id)
    res = requests.get(url, headers=headers).text
    soup = BeautifulSoup(res, "html.parser")
    if not soup.select('div.wrap_info') :
        continue;
    
    songID.append(Id)
    
    name = soup.select('div.song_name')
    name = name[0].text.replace("곡명", "").strip()
    song_name.append(name)
    print(Id)
    
    #가수가 1개가 아닐수도있음 수정필요
    singerId = soup.select('div.artist')
    if singerId[0].find("a") :
        singerId = singerId[0].find("a")["href"]
        singerId = re.findall("\d+",singerId)[0]
    else :
        singerId = singerId[0].text.strip()
    singerID.append(singerId)
    
    albumId = soup.select('div.meta > dl > dd')
    albumId = albumId[0].find("a")["href"]
    albumId= re.findall("\d+",albumId)[0]
    albumID.append(albumId)
    
    type_ = soup.select('div.wrap_atist_info > dl > dd')
    type_name = soup.select('div.wrap_atist_info > dl > dt')

    #장르 배열로 바꾸기
    genre = soup.select('div.meta > dl > dd')
    genre_name = soup.select('div.meta > dl > dt')
    position = 0
    for i in genre_name :
        if i.text == "장르" :
            break
        position+=1
    if genre[position] :
        genre = genre[position].text
    else :
        genre = []
    song_genre.append(genre)
    
    lyric_ = soup.select('div.lyric')
    if lyric_ :
        lyric_ = lyric_[0].text.strip()
    lyric.append(lyric_)
```

```python
import pandas as pd
from IPython.display import display

df = {"songID": songID, "song_name": song_name, "singerID": singerID, "albumID": albumID, "song_genre": song_genre, "lyric": lyric}
display(pd.DataFrame(df))
```
